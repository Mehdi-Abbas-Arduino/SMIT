```python
# MULTIPLE INHERITANCE
class Engine:
    def start_engine(self):
        return "Engine started."

class Radio:
    def play_music(self):
        return "Playing music."

class Car(Engine, Radio):
    def drive(self):
        return "Car is driving."

# Creating an instance of Car
my_car = Car()

# Accessing methods from both base classes
print(my_car.start_engine())  # Output: Engine started.
print(my_car.play_music())    # Output: Playing music.
print(my_car.drive())         # Output: Car is driving.
```

```python
# MULTILEVEL INHERITANCE 
class Animal:
    def eat(self):
        return "Eating food."

class Mammal(Animal):
    def give_birth(self):
        return "Giving birth."

class Dog(Mammal):
    def bark(self):
        return "Barking."

# Creating an instance of Dog
my_dog = Dog()

# Accessing methods from all levels of inheritance
print(my_dog.eat())        # Output: Eating food.
print(my_dog.give_birth()) # Output: Giving birth.
print(my_dog.bark())       # Output: Barking.
```
